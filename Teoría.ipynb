El conocimiento de los métodos para calcular las reservas de condensado y los componentes útiles asociados en un depósito de condensado de gas es necesario en el proceso de preparación de datos para la elaboración de documentos de diseño para el uso industrial de depósitos de hidrocarburos y para garantizar el desarrollo eficiente de los recursos del subsuelo.

# Justificación teórica.  

 
El gas de yacimiento de un depósito de condensado de gas es una mezcla de hidrocarburos gaseosos y líquidos de composición similar, así como gases no hidrocarburos, que son capaces de disolverse mutuamente entre sí en determinadas condiciones termobáricas.

Para determinar la composición del gas del yacimiento, es necesario que en los 2-3 pozos más productivos:
1. Se y examinar muestras de gas;
2. Se seleccione y examine muestras de condensado crudo;
3. Se mida el factor de condensado - gas (КГФ). (muestra qué tan rico es el gas en condensado, lo que es importante para determinar la rentabilidad del desarrollo.)

Si el depósito de gas tiene un borde de petróleo o una gran capa portadora de gas (>300 m), las muestras de gas y condensado deben caracterizar el cambio en el contenido de condensado crudo y la composición del gas a lo largo de la altura del depósito.   
Los cálculos utilizan los valores promedio ponderados del contenido de gases de etano, propano, butanos, C5+ por área del yacimiento.

La muestra del gas de formación se realiza mediante separación de una sola etapa si:   
La temperatura de separación es igual a la temperatura ambiente o inferior,  
    
y mediante separación de dos etapas si:   
La temperatura de separación es significativamente superior a la temperatura ambiente.



El estudio de la composición del gas de yacimiento se realiza en condiciones de laboratorio utilizando muestras de gas separado y condensado crudo.   

Las muestras de condensado crudo se someten a una desgasificación secuencial, es decir, eliminación completa de:   
metano (CH4)   
dióxido de carbono (CO2)   
nitrógeno (N2)   
sulfuro de hidrógeno (H2S)   
y eliminación parcial de homólogos de metano, y luego desbutanización, es decir, eliminación completa de los homólogos del metano (etano,propano, butano).

Para determinar el contenido de condensado, los productos de los pozos de condensado de gas se separan en condiciones de campo (separación de una o dos etapas).
El condensado crudo que cae del separador se estabiliza luego en condiciones estándar (t=20°C y P=0,1 MPa).

El contenido total de condensado estable en la composición del gas de yacimiento (en gramos-moles) se determina sumando los contenidos de:

El laboratorio determina la composición:

Además, se mide la cantidad de gas contenido en el volumen del recipiente liberado durante la desgasificación del condensado crudo (l.); la cantidad de gas liberado durante la desbutanización del condensado (l); la cantidad de hidrocarburos líquidos (C5+) en el condensado desbutanizado ($cm^3$).


# Preguntas y tareas

## Tarea 1. Cálculo de la composición del gas de yacimiento de un depósito de condensado de gas con separación de una sola etapa.

El cálculo de la composición del gas de reserva de un depósito de condensado de gas durante la separación en una sola etapa se realiza en base a 1000 gramos-moles de gas separado de acuerdo con el siguiente esquema:   

Paso 1. Determinamos el número de gramos-moles de componentes individuales del gas de separación en base a 1000 moles-gramos de gas separado:

$X_{Li} = \frac {Y_{Li} * \sum X_{Li}} {100}$   ;  $\sum X_{Li} = 1000$   

Los datos iniciales y los resultados de los cálculos se presentan en las Tabla xxx

Paso 2.  Determinamos la cantidad de gas (en gramos-moles) liberado durante la desgasificación del condensado crudo:   

$\sum X_{Ai} = \frac{a * K_g} {V} $    

donde:   
 $a$ es la cantidad de gas liberado durante la desgasificación del condensado crudo en el volumen del contenedor, l   
 $K_g$ – factor de condensado - gas (КГФ) $cm^3 / m^3$; se calcula como el cociente de dividir el volumen de condensado crudo ($cm^3$) acumulado en el separador entre el volumen de gas que ha pasado por el DICT (4 en la figura) por un mismo tiempo (a una presión en el separador no superior a 5 MPa)
 
 $K_g = \frac{V_{Ki}} {V_{gi}}$

 donde:   
 $V$ es el volumen del recipiente en el que se recogió el condensado crudo, $cm^3$   



Paso 3. Determinamos el número de moles-gramos de los componentes individuales del gas desgasificado en función del número total de moles-gramos del gas desgasificado y del contenido porcentual de los componentes en la composición del gas de yacimiento.

$X_{Ai} =  \frac{Y_{Ai} *{\sum X_{Ai} }} {100}$

Paso 4. Determinamos la cantidad de gas en gramos-moles liberados durante la desbutanización del condensado.

$\sum X_{bi} = \frac{\delta *K_g} {V}$   

donde $\delta$ : Cantidad de gas de desbutanización en el volumen del recipiente (l)

Paso 5. Determinamos el número de moles-gramos de los componentes individuales del gas desbutanizado:   

$X_{bi} =  \frac{Y_{bi} *{\sum X_{bi} }} {100}$   

Paso 6. Determinamos la cantidad (en gramos-moles) de pentanos e hidrocarburos de mayor punto de ebullición (C5+) contenidos en el condensado crudo.

$X_{vi} =  \frac{K_g*b*g_{4(k)}^{20}*24.04} {V*M_k}$   

Donde   
$b$: es el contenido de hidrocarburos líquidos (C5+) en el condensado desbunatizado en el volumen del recipiente $cm^3$   
$g_{4(k)}^2$: Densidad relativa del C5+  a los 20°C   
$M_k$: Masa molecular del C5+   
24.04: Constante de los gases

Paso 7. Determinamos el número de moles-gramos de componentes individuales del gas de yacimiento.

$X_i = X_{Li} + X_{Ai} + X_{bi} + X_{vi}$   

La cantidad de gas en gramos-moles se determina como la suma de $Xi$ y se calcula el contenido porcentual de cada componente en la composición del gas del yacimiento:   

$Y_i = \frac {X_i} {\sum X_i}  * 100$
